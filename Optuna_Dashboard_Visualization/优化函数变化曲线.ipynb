```python
import csv
import numpy as np

# 读取 random_origin.csv 文件
with open('D:/1_科研进展/小论文/小论文图片/优化目标函数变化/random_origin.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)

# 获取第1、2列并转换为数组
column_data_1 = np.array([float(row[1]) for row in data])
column_data_2 = np.array([float(row[2]) for row in data])

# 创建空的数组来存放结果
GAO_random = []
BO_random = []

# 遍历前100个元素，获取其最大值
for i in range(1, 101):
    max_value_1 = np.max(column_data_1[:i])
    max_value_2 = np.max(column_data_2[:i])
    GAO_random.append(max_value_1)
    BO_random.append(max_value_2)

# 读取 ANN_origin.csv 文件
with open('D:/1_科研进展/小论文/小论文图片/优化目标函数变化/ANN_origin.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)

# 获取第1、2列并转换为数组
column_data_1 = np.array([float(row[1]) for row in data])
column_data_2 = np.array([float(row[2]) for row in data])

# 创建空的数组来存放结果
GAO_ANN = []
BO_ANN = []

# 遍历前100个元素，获取其最大值
for i in range(1, 101):
    max_value_1 = np.max(column_data_1[:i])
    max_value_2 = np.max(column_data_2[:i])
    GAO_ANN.append(max_value_1)
    BO_ANN.append(max_value_2)

GAO_random, BO_random, GAO_ANN, BO_ANN

# 合并四个数组
final_data = list(zip(GAO_random, BO_random, GAO_ANN, BO_ANN))

# 定义新的CSV文件路径
output_file = 'D:/1_科研进展/小论文/小论文图片/优化目标函数变化/final.csv'

# 写入新的CSV文件
with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['GAO_random', 'BO_random', 'GAO_ANN', 'BO_ANN'])  # 写入表头
    writer.writerows(final_data)  # 写入数据

print(f'文件已保存至 {output_file}')
```
